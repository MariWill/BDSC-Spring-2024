```python
def write_srtype_counts_to_file(df, filename = 'srtypes.txt'):    
    srtype_counts = df['SRType'].value_counts()
    # Open a text file for writing
    with open(filename, 'w') as file:
        # Write the counts for each 'srtypes' with line numbers
        print("#", " SRType", "                 Counts", file=file)
        for i, (srtype, count) in enumerate(srtype_counts.items(), 1):
            print(f"{i}. {srtype}: {count} occurrences", file=file)
```
